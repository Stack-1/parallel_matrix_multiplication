```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np

kij = []
ikj = []
df = pd.read_csv('../../../../../../times.csv')

for i in df['6']:
    kij.append(float(i))
    
df = pd.read_csv('../../../../../../times_2.csv')
for i in df['6']:
    ikj.append(float(i))

y = []

for i in range(0,len(df['3'])):
    y.append(str(df['1'][i]) + 'x' + str(df['2'][i]) + 'x' + str(df['3'][i]))
print(y)
    
fig, ax = plt.subplots()
ax.plot(np.array(kij[0:13]),label='kij')
ax.plot(np.array(ikj[0:13]),label='ikj')
ax.set_xticklabels(y[0:13], rotation=45)

ax.set(xlabel='', ylabel='time (s)',
       title='Comparison between ikj and kij (Square Matrices)')
ax.grid()
ax.legend()
plt.show()
```
